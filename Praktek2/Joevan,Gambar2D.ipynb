```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk membuat lingkaran midpoint dengan titik-titik
def draw_circle_midpoint(xc, yc, r):
    x = 0
    y = r
    p = 1 - r

    x_points = []
    y_points = []

    while x <= y:
        # 8-way symmetry
        x_points.extend([xc + x, xc - x, xc + x, xc - x, xc + y, xc - y, xc + y, xc - y])
        y_points.extend([yc + y, yc + y, yc - y, yc - y, yc + x, yc + x, yc - x, yc - x])
        x += 1
        if p < 0:
            p += 2 * x + 1
        else:
            y -= 1
            p += 2 * (x - y) + 1

    return x_points, y_points

# Setup plot
plt.figure()
plt.axis('equal')
plt.grid(True)
plt.xlim(-70, 70)
plt.ylim(-50, 50)

# Lingkaran biru (sebagai titik penuh)
circle_biru = plt.Circle((-55, -20), 8, color='navy')
plt.gca().add_patch(circle_biru)

# Lingkaran merah (midpoint)
x_red, y_red = draw_circle_midpoint(-25, -15, 10)
plt.scatter(x_red, y_red, s=10, color='r', label='Red Circle')

# Lingkaran hijau besar (parametrik, garis putus-putus)
theta = np.linspace(0, 2 * np.pi, 360)
r_hijau = 40
x_hijau = r_hijau * np.cos(theta) - 5  # digeser ke kiri sedikit
y_hijau = r_hijau * np.sin(theta) - 10
plt.plot(x_hijau, y_hijau, 'g--', label='Green Circle')

# Lingkaran ungu besar (midpoint)
x_ungu, y_ungu = draw_circle_midpoint(20, 0, 50)
plt.scatter(x_ungu, y_ungu, s=5, color='purple', label='Purple Circle')

# Panah sumbu x dan y
plt.arrow(0, 0, 30, 0, head_width=2, head_length=4, fc='k', ec='k')  # x arrow
plt.arrow(0, 0, 0, -30, head_width=2, head_length=4, fc='k', ec='k')  # y arrow
plt.text(32, 0, 'x', fontsize=12)
plt.text(0, -32, 'y', fontsize=12)

# Tampilkan semua
plt.title("Rekonstruksi Gambar Lingkaran")
plt.legend()
plt.show()
```

```python
# Listing 2.7: Program ELLIPSES

"""
ELLIPSES
"""

import numpy as np
import matplotlib.pyplot as plt

plt.axis([-75, 75, -50, 50])
plt.grid(True)

plt.arrow(0, 0, 60, 0, head_length=4, head_width=3, color='k')
plt.arrow(0, 0, 0, -40, head_length=4, head_width=3, color='k')

plt.text(65, 0, 'x')
plt.text(5, -44, 'y')

# ========================= RED ELLIPSE =========================
a = 40
b = 20
p = 2 * np.pi / 180

xplast = a * np.cos(0)
yplast = b * np.sin(0)

for tp in np.arange(p, 2 * np.pi + p, p):
    xp = a * np.cos(tp)
    yp = b * np.sin(tp)
    plt.plot([xplast, xp], [yplast, yp], color='r')
    xplast = xp
    yplast = yp

# ========================= GREEN ELLIPSE =========================
a = 30
b = 20
p = 2 * np.pi / 180

xplast = a * np.cos(0)
yplast = b * np.sin(0)

for tp in np.arange(p, 2 * np.pi + p, p):
    xp = a * np.cos(tp)
    yp = b * np.sin(tp)
    plt.plot([xplast, xp], [yplast, yp], linestyle='--', color='g')
    xplast = xp
    yplast = yp

# ========================= BLUE ELLIPSE =========================
a = 60
b = 30
p = 2 * np.pi / 180

xplast = a * np.cos(0) - 5
yplast = b * np.sin(0) - 5

for tp in np.arange(p, 2 * np.pi + p, p):
    xp = a * np.cos(tp) - 5
    yp = b * np.sin(tp) - 5
    plt.plot([xplast, xp], [yplast, yp], color='b')
    xplast = xp
    yplast = yp

plt.show()
```

```python
"""
2DTRANSLATION
"""

import numpy as np
import matplotlib.pyplot as plt

x1=-10
x2=140
y1=90
y2=-10
plt.axis([x1,x2,y1,y2])

plt.axis('on')
plt.grid(True)
plt.title('Translation')

# _________________________________triangle
x=[20,30,40,20]
y=[40,20,40,40]
plt.plot(x,y,color='k')
plt.plot(x,y,color='k')
plt.plot(x,y,color='k')

# _____________________________translate triangle dx=60
x=[60,70,80,60]
plt.plot(x,y,color='g')
plt.plot(x,y,color='g')
plt.plot(x,y,color='g')

# _____________________________translate triangle dy=40
y=[80,60,80,80]
plt.plot(x,y,color='r')
plt.plot(x,y,color='r')
plt.plot(x,y,color='r')
# _____________________________________________box
x=[0,0,5,5,0]
y=[55,50,50,55,55]
plt.plot(x,y,'b')

# ________________________________translate box
y=[55,50,50,55,55]
for x in np.arange(0,130,10):
    x=[x,x,x+5,x+5,x]
    plt.plot(x,y,'b')

plt.show()
```

```python
# Listing 2.11 - Program 2DROTRECTANGLE

"""
2DROTRECTANGLE
"""

import matplotlib.pyplot as plt
import numpy as np

plt.axis([-100, 150, -100, 100])
plt.axis('on')
plt.grid(True)

# ======================== AXES ========================
plt.arrow(0, 0, 60, 0, head_length=5, head_width=3, color='k')
plt.arrow(0, 0, 0, 60, head_length=5, head_width=3, color='k')
plt.text(65, 0, 'x')
plt.text(0, 65, 'y')

# ======================== CENTER OF ROTATION ========================
xc = 10
yc = 20
plt.plot([0, xc], [0, yc], linestyle='dashed', color='grey')  # Garis bantu ke pusat rotasi
plt.text(xc + 3, yc + 3, 'C')
plt.scatter([xc], [yc], color='k')  # Titik pusat rotasi

# ======================== FUNGSI ROTASI ========================
def rotate(x, y, xr, yr, t):
    xr = x - xr
    yr = y - yr
    x_rot = xr * np.cos(t) - yr * np.sin(t)
    y_rot = xr * np.sin(t) + yr * np.cos(t)
    return [x_rot, y_rot]

# ======================== RECTANGLE (Sebelum Rotasi) ========================
x0 = 40
y0 = 60
x1 = 80
y1 = 60
x2 = 80
y2 = 90
x3 = 40
y3 = 90
x4 = x0
y4 = y0

# Gambar persegi panjang sebelum rotasi
plt.plot([x0, x1, x2, x3, x4], [y0, y1, y2, y3, y4], 'k')  # warna hitam

# Koordinat dalam sistem relatif terhadap pusat rotasi
xp0 = x0 - xc
yp0 = y0 - yc
xp1 = x1 - xc
yp1 = y1 - yc
xp2 = x2 - xc
yp2 = y2 - yc
xp3 = x3 - xc
yp3 = y3 - yc
xp4 = xp0
yp4 = yp0

# ======================== ROTASI ========================
t = np.pi / 6  # 30 derajat dalam radian

# Titik setelah rotasi
[xr0, yr0] = rotate(x0, y0, xc, yc, t)
[xr1, yr1] = rotate(x1, y1, xc, yc, t)
[xr2, yr2] = rotate(x2, y2, xc, yc, t)
[xr3, yr3] = rotate(x3, y3, xc, yc, t)
[xr4, yr4] = rotate(x4, y4, xc, yc, t)

# Plot hasil rotasi
plt.plot([xr0, xr1, xr2, xr3, xr4], [yr0, yr1, yr2, yr3, yr4], color='r')  # warna merah

# Tampilkan hasil
plt.show()
```
